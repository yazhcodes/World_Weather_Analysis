```python
min_temp = float(input("How much cold can you handle?"))
max_temp = float(input("How much heat can you handle?"))
print('Here you go! You might love these vacation destinations..')
vaca_df = city_df.loc[
                    (city_df['Max Temp'] >= min_temp) & 
                    (city_df['Max Temp'] <= max_temp)
                    ].dropna().head(200)
```

```python
hotel_df = vaca_df[['City','Country','Lat','Lon','Max Temp']]
hotel_df['Hotel Name'] = ''
hotel_df = hotel_df.reset_index().drop(columns = 'index')
hotel_df
```

```python
base_url = 'https://maps.googleapis.com/maps/api/place/nearbysearch/json?'
params = {
    'type' : 'lodging',
    'keyword' : 'hotel',
    'radius' : 5000,
    'key' : g_key
}
result_count = len(hotel_df)
for i in range(0,result_count):
    params['location'] = f"{hotel_df['Lat'][i]},{hotel_df['Lon'][i]}"
    try:
        response = requests.get(base_url,params)
        hotel_df['Hotel Name'][i] = response.json()['results'][0]['name']
        print(f"Successfully Processed {i+1} of {result_count}")
    except:
        print(f"Processing Failed for {i+1} of {result_count}")
```

```python
hotel_df
```

```python
vaca_locations = hotel_df[['Lat','Lon']]
heatmap_layer = gmaps.heatmap_layer(
                        vaca_locations, 
                        hotel_df['Max Temp'],
                        dissipating=False, 
                        max_intensity=300, 
                        point_radius=4
                        )
info_box_template = """
<dl>
    <dt>Nearest Hotel</dt><dd>{Hotel Name}</dd>
    <dt>City</dt><dd>{City}</dd>
    <dt>Country</dt><dd>{Country}</dd>
    <dt>Max Temp</dt><dd>{Max Temp}</dd>
</dl>
"""
hotel_info = [info_box_template.format(**i) for i in hotel_df.to_dict('records')]
marker_layer = gmaps.marker_layer(vaca_locations, info_box_content=hotel_info)

fig = gmaps.figure(
    center = (30.0, 31.0),
    zoom_level = 1.75
    )
fig.add_layer(heatmap_layer)
fig.add_layer(marker_layer)
fig
```
