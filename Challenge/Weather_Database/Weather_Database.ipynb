```python
# Import Dependencies
import random
from citipy import citipy
import requests
from config import weather_api_key 
import pandas as pd
```

```python
# Initiate Lists
cities = []
country = []
lat = []
lng = []
max_temp = []
humidity = []
cloudiness = []
wind_speed = []
current_desc = []
city_count = 1
```

```python
# Iterate 2000 times
for i in range(0,2000):
    
    # Generate random latitude and longitude
    latitude = random.uniform(-90,90)
    longitude = random.uniform(-180,180)
    
    # Find nearest city for each random coordinate
    city_name = citipy.nearest_city(latitude,longitude).city_name
    country_code = citipy.nearest_city(latitude,longitude).country_code
      
    # Fetch Weather Data for each City
    url = f'http://api.openweathermap.org/data/2.5/weather?units=imperial&q={city_name},{country_code}&appid={weather_api_key}'
    response = requests.get(url).json()
    try:
        if response['name'] not in cities:
            cities.append(response['name'])
            country.append(response['sys']['country'])
            lat.append(response['coord']['lat'])
            lng.append(response['coord']['lon'])
            max_temp.append(response['main']['temp_max'])
            humidity.append(response['main']['humidity'])
            cloudiness.append(response['clouds']['all'])
            wind_speed.append(response['wind']['speed'])
            current_desc.append(response['weather'][0]['description'])
            print(f'{city_count}: Weather Data retrieved for {cities[-1]}')
            city_count += 1
    except:
        print(f'Weather Data unavailable for {city_name}')
```

```python
city_weather_df = pd.DataFrame({
                        'City' : cities,
                        'Country' : country,
                        'Lat' : lat,
                        'Lng' : lng,
                        'Max Temp' : max_temp,
                        'Humidity' : humidity,
                        'Cloudiness' : cloudiness,
                        'Wind Speed' : wind_speed,
                        'Current Description' : current_desc
                        })
city_weather_df
```

```python
city_weather_df.to_csv('WeatherPy_Database.csv')
```
